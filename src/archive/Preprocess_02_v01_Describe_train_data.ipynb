# Describe train data

This notebook describes the train data in a designated folder

```python
# Import necessary libraries
import os
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Function to count .tif files in each subdirectory
def count_tif_files(directory):
    subdirectory_counts = {}
    for subdir, _, files in os.walk(directory):
        count = len([file for file in files if file.endswith('.tif')])
        if count > 0 and subdir != directory:  # Exclude the main directory itself
            subdirectory_name = os.path.basename(subdir)  # Get only the subdirectory name
            subdirectory_counts[subdirectory_name] = count
    return subdirectory_counts
```

```python
# Path to the main directory (update this with the actual path)
main_directory = 'data/DETAILED_merged'
```

```python
# Initialize a counter for .tif files
tif_count = 0

# Traverse the directory structure
for root, dirs, files in os.walk(main_directory):
    # Count .tif files in the current directory
    tif_count += len([file for file in files if file.endswith('.tif')])

# Display the total number of .tif files
print(f"Total number of images in {main_directory}: {tif_count}")
```

```python
# Get the counts of .tif files in each subdirectory
tif_counts = count_tif_files(main_directory)

# Create a DataFrame from the counts for better visualization and manipulation
df = pd.DataFrame(list(tif_counts.items()), columns=['Label', 'Count'])

# Sort the DataFrame by the 'Count' column in descending order
df = df.sort_values(by='Count', ascending=False)
```

```python
# Plot the bar chart with logarithmic scale on the y-axis
plt.figure(figsize=(10, 6))
plt.bar(df['Label'], df['Count'], color='skyblue')
plt.yscale('log')
plt.xlabel('Class')
plt.ylabel('Number of images')
plt.title('Count of images per Class')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to make room for x-axis labels
plt.show()
```
