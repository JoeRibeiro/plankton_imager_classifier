# Combining outputs & treshold

This notebook provides code to combine various output .csv's with predictions, and provide a way to set a treshold for a quick evaluation of the models confidence.

```python
import glob
import os
import pandas as pd

#for tensor
import fastai
from fastai.vision.all import *
```

```python
%%time 

version = 'v03'
folder_name = 'preds/Preds_new_data/'+version
file_type = 'csv'
seperator =','
```

```python
dataframe = pd.concat([pd.read_csv(f, sep=seperator) for f in glob.glob(folder_name + "/*."+file_type)],ignore_index=True)
```

```python
%%time

dataframe
#dataframe.head()
```

```python
%%time

# Change to tensor and extract highest value
dataframe_tensor = torch.tensor(dataframe.iloc[:,2:].to_numpy())
out, inds = torch.max(dataframe_tensor,dim=1)
```

```python
%%time

# For dataframe collected from separate files (see above)

# Evaluate how many preds are above treshold

# Set threshold value
threshold = .98

# Create tensor with True/False for every value in y
b = out > threshold

# Calculate number of images above threshold
ImageNr = len(dataframe_tensor[b]) 
# Calculate percentage of images above threshold
ImagePerc = (len(dataframe_tensor[b])*100/len(dataframe_tensor))

print (ImageNr, 'images (i.e.,', float("{:.2f}".format(ImagePerc)), 
       '%) of predictions are above the threshold of', threshold)
print()
```

```python
%%time

# Add column number and highest value of preds to table
dataframe['out'] = out
dataframe['inds'] = inds
```

```python
%%time

dataframe.head()
```

```python
%%time

# Export table to .csv file for downstream applications
# Reduced decimal points by '%.3f'
dataframe.to_csv(version+'_All_data.csv', index=False, float_format='%.3f') 
```
